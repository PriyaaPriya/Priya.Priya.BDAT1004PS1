```python
Question 3 C#
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area 
of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c)
, where 
s = (a+b+c)/2.

Answer 3:
public static double triangleArea(int a, int b, int c)
        {
            double s = (a + b + c) / 2.0;
            double har = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return har;
        }
```

```python
Question 7 Python
Write a function bldcount() that reads the file with name name and reports (i.e., prints) how many patients there are in each bloodtype

Answer 7:

import re
def bldcount(ff):
    try:
        myFile=open(ff,"r")
        A=0
        B=0
        AB=0
        O=0
        OO=0
        bt=myFile.readlines()
        for line in bt:
            if re.search("A", line):
               A=A+1
            elif re.search("B", line):
                B=B+1
            elif re.search("AB", line):
                AB=AB+1
            elif re.search("O", line):
                O=O+1
            elif re.search("OO", line):
                OO=OO+1
        print("there are "+A+" patients of blood type A.")
        print("There is "+B+" patient of blood type B.")
        print("there are " + AB + " patients of blood type AB.")
        print("there are " + O + " patients of blood type O.")
        print("there are " + OO + " patients of blood type OO.")


    except:
        print("Not able to retrieve file")
    finally:
        myFile.close()
bldcount('bloodtype.txt')

```

```python
Question 8 Python
Write a function curconv() that takes as input:

Answer 8:
def curconv(crs, cr):
    try:
        myFile=open("currencies.txt","r")
        crInfo=myFile.readlines()
        for line in crInfo:
            fields = line.split(' ')
            if fields[0]==crs:
                print(float(fields[1])*cr)
    except:
        print("Not able to retrieve file")
    finally:
        myFile.close()

curconv('EUR', 100)

```
